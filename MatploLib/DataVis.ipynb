# Data Visualisation

## Matplotlib

- The foundation for all plots
- like the C language of visualisation
- seaborn and pandas plot are built on it

```python
import matplotlib.pyplot as plt
```

```python
epochs = [1,2,3,4,5]
train_loss = [0.9,0.6,0.5,0.4,0.3]
val_loss = [1,0.7,0.6,0.5,0.35]

plt.plot(epochs, train_loss, label = "Training Loss")
plt.plot(epochs, val_loss, label = "Validation Loss")
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.title('Training vs Validation Loss')
plt.legend()
plt.grid(True)
plt.show()

```

```python
subjects = ["math","science","social"]
marks = [20,30,40]

plt.bar(subjects, marks)
```

```python
import seaborn as sns
df = sns.load_dataset("iris")

plt.hist(df['sepal_length'], bins = 20)
plt.title('Distribution of Sepal Length')
plt.xlabel('Sepal Length')
plt.ylabel('Count')
plt.show()
```

## Seaborn Library

```python
import seaborn as sns

df = sns.load_dataset('titanic')
```

```python
df.head()
```

```python
sns.countplot(x = "survived", data = df)
plt.title('Survivors in Titanic')
plt.xlabel('Survived (0 = No, 1 = Yes)')
plt.ylabel('Count')
plt.show()
```

```python
sns.boxplot(x = "survived", y = 'age', data = df)

```
